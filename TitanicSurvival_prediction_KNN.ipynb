```python
passenger_id_df = test_df['PassengerId']
```

```python
columns_to_drop = ['Cabin', 'PassengerId', 'Name', 'Ticket']

train_df = train_df.drop(columns=columns_to_drop)

test_df = test_df.drop(columns=columns_to_drop)
```

```python
train_df.head()
```

```python
age_mean = train_df['Age'].mean()

print(age_mean)
```

```python
train_df['Age'] = train_df['Age'].fillna(age_mean)
```

```python
test_df['Age'] = test_df['Age'].fillna(age_mean)
```

```python
fare_mean = train_df['Fare'].mean()

print(fare_mean)
```

```python
test_df['Fare'] = test_df['Fare'].fillna(fare_mean)
```

```python
embarked_mode = train_df['Embarked'].mode()[0]

print(embarked_mode)
```

```python
train_df['Embarked'] = train_df['Embarked'].fillna(embarked_mode)
```

```python
print(train_df.isna().sum())
```

```python
print(test_df.isna().sum())
```

```python
train_df = pd.get_dummies(train_df)

test_df = pd.get_dummies(test_df)
```

```python
train_df.head()
```

```python
y = train_df['Survived']
X = train_df.drop('Survived', axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X = scaler.fit_transform(X)
```

```python
test_df = scaler.transform(test_df)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
```

```python
knn = KNeighborsClassifier()
```

```python
param_grid = {
    'n_neighbors': list(range(1, 31)),
    'p': [1, 2]
}
```

```python
grid_search = GridSearchCV(
    estimator=knn,
    param_grid=param_grid,
    cv=5,
    scoring='accuracy'
)
```

```python
grid_search.fit(X, y)
```

```python
print("Best Parameters:")
print(grid_search.best_params_)

print("\nBest Accuracy:")
print(grid_search.best_score_)
```

```python
best_model = grid_search.best_estimator_
```

```python
predicted_test_Y = best_model.predict(test_df)
predicted_test_Y[:10]
```

```python
predicted_test_Y_df = pd.DataFrame()

predicted_test_Y_df['PassengerId'] = passenger_id_df
predicted_test_Y_df['Survived'] = predicted_test_Y
```

```python
predicted_test_Y_df.to_csv(
    "titanic_knn_submission.csv",
    index=False,
    header=True
)
```
